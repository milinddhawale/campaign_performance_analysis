### For the next steps, I will:

- Investigate the performance of active campaigns.
- Analyze the distribution of campaigns across Branch and CCC.
- Identify any correlations or patterns that might be relevant for business insights.

I will begin by filtering out active campaigns and then analyze their presence across branches and ccc.

```python
# Filter the active campaign

active_campaign_data = df[df['STATUS'] == 'Active']

# count the active campaign in the branch and ccc

branch_distribution = active_campaign_data['Branch'].value_counts()
ccc_distribution = active_campaign_data['CCC'].value_counts()

# display the distribution
print("count of active campaign in Branch Distribution",branch_distribution)
print("count of active campaign in CCC Distribution(Customer care center)", ccc_distribution)
```
